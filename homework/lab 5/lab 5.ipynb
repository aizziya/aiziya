```python
#1
class User:
    def __init__(self, id, name, email):
        self._id = id
        self._name = ' '.join(name.strip().split()).title()
        self._email = email.strip().lower()
        if '@' not in self._email:
            raise ValueError('Email address is not valid')
    def __str__(self):
        return (f"User(id={self._id}, name='{self._name}', email='{self._email})'")
    def __del__(self):
        print(f"User {self._name} deleted")
#2
    @classmethod
    def from_string(cls, data: str):
        parts = [p.strip() for p in data.split(',')]
        _id = int(parts[0].strip())
        _name = parts[1].strip()
        _email = parts[2].strip()
        return cls(_id, _name, _email)
u = User.from_string("2, Alice Wonderland , alice@wonder.com")
print(u)
#3
class Product:
    def __init__(self, id, name, price, category):
        self.id = id
        self.name = name.strip().title()
        self.price = float(price)
        self.category = category
    def __str__(self):
        return(f"Product(id={self.id}, name='{self.name}', price={self.price}, category='{self.category}')")
    def __hash__(self):
        return hash(self.id)
    def __eq__(self, other):
        if not isinstance(other, Product):
            return NotImplemented
        return self.id == other.id
    def to_dict(self):
        return {'id': self.id, 'name': self.name, 'price': self.price, 'category': self.category}
#4
class Inventory:
    def __init__(self):
        self._products: Dict[int, Product] = {}

    def add_product(self, product: Product):
        self._products[product.id] = product

    def remove_product(self, product_id: int):
        if product_id in self._products:
            del self._products[product_id]

    def get_product(self, product_id: int) -> Product:
        return self._products.get(product_id)

    def get_all_products(self) -> List[Product]:
        return list(self._products.values())

    def unique_products(self) -> Set[Product]:
        return set(self._products.values())

    def to_dict(self) -> Dict[int, Product]:
        return self._products.copy()

```
