```python
#1
def invert_unique(d):
    result = {}

    for key, value in d.items():
        if value not in result:
            result[value] = []
        if key not in result[value]:
            result[value].append(key)

    return result

d = {'a': 1, 'b': 2, 'c': 1, 'd': 3, 'e': 2}
print("Результат:", invert_unique(d))
```

```python
#3
def merge_dicts_sum(d1, d2):
    result = {}

    for key, value in d1.items():
        result[key] = value

    for key, value in d2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value

    return result

d1 = {'a': 5, 'b': 3, 'c': 2}
d2 = {'b': 4, 'c': 1, 'd': 7}
print("Результат:", merge_dicts_sum(d1, d2))
```

```python
#5
sort_dict_by_values = lambda d: sorted(d.keys(), key=lambda k: (-d[k], k))[:5]

d = {'apple': 10, 'banana': 5, 'cherry': 15, 'date': 10, 'elderberry': 3, 'fig': 8}
print("Результат:", sort_dict_by_values(d))
```

```python
#6
def deep_sum(d):
    total = 0

    for value in d.values():
        if isinstance(value, (int, float)):
            total += value
        elif isinstance(value, list):
            for num in value:
                if isinstance(num, (int, float)):
                    total += num
        elif isinstance(value, dict):
            total += deep_sum(value)

    return total

d = {'a': 5, 'b': [1, 2, 3], 'c': {'x': 10, 'y': [4, 5]}, 'd': 7}
print("Результат:", deep_sum(d))
```

```python
#8
def sort_dict_by_value_length(d):
    items = list(d.items())

    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            if len(items[i][1]) > len(items[j][1]):
                items[i], items[j] = items[j], items[i]
            elif len(items[i][1]) == len(items[j][1]) and items[i][0] > items[j][0]:
                items[i], items[j] = items[j], items[i]

    return items

d = {'cat': 'meow', 'dog': 'woof', 'bird': 'tweet', 'cow': 'moo'}
print("Результат:", sort_dict_by_value_length(d))
```

```python
#10
filter_dict_odd_sorted = lambda d: {k: sorted([x for x in v if x % 2 == 1])
                                   for k, v in d.items()
                                   if len([x for x in v if x % 2 == 1]) > 0}

d = {'a': [1, 2, 3, 4], 'b': [2, 4, 6], 'c': [5, 7, 9], 'd': [1, 3, 5]}
print("Результат:", filter_dict_odd_sorted(d))
```

```python
#11
def group_by_length(words):
    result = {}
    seen = {}

    for word in words:
        length = len(word)
        if length not in result:
            result[length] = []
            seen[length] = set()

        if word not in seen[length]:
            seen[length].add(word)
            result[length].append(word)

    return result

words = ['cat', 'dog', 'elephant', 'bird', 'cat', 'lion', 'tiger']
print("Результат:", group_by_length(words))
```

```python
#13
def invert_dict_strict(d):
    value_count = {}
    for value in d.values():
        if value in value_count:
            value_count[value] += 1
        else:
            value_count[value] = 1

    result = {}
    for key, value in d.items():
        if value_count[value] == 1:
            result[value] = key

    return result

d = {'a': 1, 'b': 2, 'c': 1, 'd': 3}
print("Результат:", invert_dict_strict(d))
```

```python
#15
filter_dict_mean_odd = lambda d: {k: v for k, v in d.items()
                                 if v > (sum(d.values()) / len(d)) and v % 2 == 1}

d = {'a': 10, 'b': 5, 'c': 8, 'd': 3, 'e': 12}
print("Результат:", filter_dict_mean_odd(d))
```

```python
#16
def update_counts(d, items):
    for item in items:
        if item in d:
            d[item] += 1
        else:
            d[item] = 1
    return d

d = {'apple': 3, 'banana': 2}
items = ['apple', 'banana', 'apple', 'cherry', 'banana']
print("Результат:", update_counts(d, items))
```

```python
#18
def sort_dict_by_value_sum(d):
    sums = {}
    for key, values in d.items():
        total = 0
        for num in values:
            total += num
        sums[key] = total

    items = list(sums.items())

    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            if items[i][1] < items[j][1]:
                items[i], items[j] = items[j], items[i]
            elif items[i][1] == items[j][1] and items[i][0] > items[j][0]:
                items[i], items[j] = items[j], items[i]

    return items

d = {'a': [1, 2, 3], 'b': [4, 5], 'c': [1, 1, 1], 'd': [10]}
print("Результат:", sort_dict_by_value_sum(d))
```

```python
#21
def count_leaf_values(d):
    count = 0

    for value in d.values():
        if isinstance(value, (int, float)):
            count += 1
        elif isinstance(value, list):
            count += len(value)
        elif isinstance(value, dict):
            count += count_leaf_values(value)

    return count

d = {'a': 5, 'b': [1, 2, 3], 'c': {'x': 10, 'y': [4, 5], 'z': {'w': 7}}}
print("Результат:", count_leaf_values(d))
```

```python
#23
def group_by_last_letter(words):
    result = {}
    seen = {}

    for word in words:
        last = word[-1]
        if last not in result:
            result[last] = []
            seen[last] = set()

        if word not in seen[last]:
            seen[last].add(word)
            result[last].append(word)

    return result

words = ['cat', 'dog', 'bird', 'bat', 'rat', 'cat', 'frog']
print("Результат:", group_by_last_letter(words))
```

```python
#25
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

filter_prime_oddkey = lambda d: {k: v for k, v in d.items()
                               if is_prime(v) and len(k) % 2 == 1}

d = {'a': 7, 'bb': 11, 'c': 4, 'ddd': 13, 'e': 5}
print("Результат:", filter_prime_oddkey(d))
```

```python
#27
filter_by_last_digit = lambda d: sorted(d.keys(), key=lambda k: (d[k] % 10, k))

d = {'a': 12, 'b': 25, 'c': 18, 'd': 23, 'e': 11}
print("Результат:", filter_by_last_digit(d))
```

```python
#29
filter_unique_strings = lambda d: {k: v for k, v in d.items()
                                 if len(set(v)) == len(v) and all(len(s) > 3 for s in v)}

d = {'a': ['hello', 'world'], 'b': ['hi', 'bye'], 'c': ['test', 'test'], 'd': ['longer', 'strings']}
print("Результат:", filter_unique_strings(d))
```

```python
#31
filter_by_avg = lambda d: {k: v for k, v in d.items()
                         if sum(v) / len(v) > (sum(sum(v) for v in d.values()) /
                                               sum(len(v) for v in d.values()))}

d = {'a': [1, 2, 3], 'b': [10, 20], 'c': [5, 5, 5], 'd': [100]}
print("Результат:", filter_by_avg(d))
```

```python
#33
filter_not_multiple3_evenkey = lambda d: {k: v for k, v in d.items()
                                        if v % 3 != 0 and len(k) % 2 == 1}

d = {'a': 7, 'bb': 9, 'c': 12, 'ddd': 5, 'e': 8}
print("Результат:", filter_not_multiple3_evenkey(d))
```

```python
#35
def factorial(n):
    if n <= 1:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

replace_with_factorial = lambda d: {k: factorial(v) if v < 6 else v for k, v in d.items()}

d = {'a': 3, 'b': 7, 'c': 5, 'd': 10, 'e': 2}
print("Результат:", replace_with_factorial(d))
```

```python
#37
def count_vowels(s):
    vowels = 'aeiou'
    count = 0
    for c in s.lower():
        if c in vowels:
            count += 1
    return count

sort_by_vowels_value = lambda d: sorted(d.keys(), key=lambda k: (count_vowels(k), -d[k]))

d = {'apple': 10, 'banana': 5, 'cherry': 8, 'date': 3, 'fig': 7}
print("Результат:", sort_by_vowels_value(d))
```
