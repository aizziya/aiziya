```python
#16
def transform_list(nums):
    result = []

    for num in nums:
        if num < 0:
            continue
        if num % 2 == 0:
            result.append(num ** 2)
        elif num > 10:
            sum_digits = 0
            temp = num
            while temp > 0:
                sum_digits += temp % 10
                temp //= 10
            result.append(sum_digits)
        else:
            result.append(num)

    return result

nums = [5, -3, 12, 8, 15, -1, 3, 20, 11]
print("Результат:", transform_list(nums))
```

```python
#17
process_numbers = lambda nums: list(map(
    lambda x: x ** 2,
    filter(
        lambda x: (x % 3 == 0 or x % 5 == 0) and x % 15 != 0 and len(str(abs(x))) % 2 == 1,
        nums
    )
))

nums = [3, 5, 15, 7, 9, 10, 30, 100, 12, 25]
print("Результат:", process_numbers(nums))
```

```python
#18
def flatten_and_filter(lst):
    numbers = []

    def extract_numbers(item):
        if isinstance(item, list):
            for sub_item in item:
                extract_numbers(sub_item)
        else:
            numbers.append(item)

    extract_numbers(lst)

    result = []
    for num in numbers:
        if num > 0 and num % 4 != 0 and len(str(num)) > 1:
            result.append(num)

    for i in range(len(result)):
        for j in range(i + 1, len(result)):
            if result[i] > result[j]:
                result[i], result[j] = result[j], result[i]

    return result

lst = [1, [2, [16, 4, -5]], 8, [12, [24, 3]], 100]
print("Результат:", flatten_and_filter(lst))
```

```python
#20
def max_subarray_sum(nums, k):
    if len(nums) < k:
        return None

    max_sum = None

    for i in range(len(nums) - k + 1):
        subarray = nums[i:i+k]

        valid = True
        for num in subarray:
            if num <= 0:
                valid = False
                break

        if valid:
            current_sum = 0
            for num in subarray:
                current_sum += num

            if max_sum is None or current_sum > max_sum:
                max_sum = current_sum

    return max_sum

nums = [3, 5, 2, 7, 1, 8, 4, 6]
k = 3
print("Результат:", max_subarray_sum(nums, k))
```

```python
#21
filter_strings = lambda strings: [
    s.upper() for s in strings
    if s.isalpha() and len(s) > 4 and len(set(s)) == len(s)
]

strings = ["hello", "world", "python", "abcde", "test", "book", "unique"]
print("Результат:", filter_strings(strings))
```

```python
#22
def group_by_parity_and_sort(nums):
    evens = []
    odds = []

    for num in nums:
        if num % 2 == 0:
            evens.append(num)
        else:
            odds.append(num)

    for i in range(len(evens)):
        for j in range(i + 1, len(evens)):
            if evens[i] > evens[j]:
                evens[i], evens[j] = evens[j], evens[i]

    for i in range(len(odds)):
        for j in range(i + 1, len(odds)):
            if odds[i] > odds[j]:
                odds[i], odds[j] = odds[j], odds[i]

    return evens + odds

nums = [5, 2, 7, 4, 1, 8, 3, 6]
print("Результат:", group_by_parity_and_sort(nums))
```

```python
#23
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

filter_by_prime_index = lambda nums: [
    nums[i] for i in range(len(nums))
    if is_prime(i) and nums[i] % 2 == 1 and nums[i] > (sum(nums) / len(nums))
]

nums = [10, 15, 21, 8, 13, 7, 9, 11, 4]
print("Результат:", filter_by_prime_index(nums))
```

```python
#24
def longest_increasing_sublist(nums):
    if not nums:
        return []

    longest = [nums[0]]
    current = [nums[0]]

    for i in range(1, len(nums)):
        if nums[i] > nums[i-1]:
            current.append(nums[i])
        else:
            if len(current) > len(longest):
                longest = current
            current = [nums[i]]

    if len(current) > len(longest):
        longest = current

    return longest

nums = [1, 2, 3, 1, 2, 3, 4, 5, 1, 2]
print("Результат:", longest_increasing_sublist(nums))
```

```python
#25
average_of_lists = lambda lists: [
    sum(inner) / len(inner) for inner in lists
    if len(inner) >= 3 and sum(inner) % 2 == 0
]

lists = [[1, 2, 3], [4, 5, 6, 7], [2, 4, 6], [1, 3, 5], [8, 10, 12]]
print("Результат:", average_of_lists(lists))
```

```python
#26
def remove_duplicates_keep_last(nums):
    seen = set()
    result = []

    for i in range(len(nums) - 1, -1, -1):
        if nums[i] not in seen:
            seen.add(nums[i])
            result.insert(0, nums[i])

    return result

nums = [1, 2, 3, 2, 4, 1, 5, 3]
print("Результат:", remove_duplicates_keep_last(nums))
```

```python
#27
sort_strings = lambda strings: sorted(
    strings,
    key=lambda s: (-len(s), s)
)[:5]

strings = ["python", "java", "javascript", "c", "c++", "ruby", "php", "swift"]
print("Результат:", sort_strings(strings))
```

```python
#28
def moving_average(nums, k):
    if len(nums) < k:
        return []

    averages = []

    for i in range(len(nums) - k + 1):
        window = nums[i:i+k]

        has_negative = False
        for num in window:
            if num < 0:
                has_negative = True
                break

        if not has_negative:
            avg = sum(window) / k
            averages.append(avg)

    return averages

nums = [1, 2, 3, 4, 5, -1, 6, 7, 8]
k = 3
print("Результат:", moving_average(nums, k))
```

```python
#29
find_unique_greater = lambda list1, list2: [
    x for x in list1
    if x not in list2 and x > (sum(list1) / len(list1))
]

list1 = [10, 15, 20, 25, 30]
list2 = [15, 25, 35, 45]
print("Результат:", find_unique_greater(list1, list2))
```

```python
#30
def analyze_strings_list(words):
    result = []
    seen = set()

    for word in words:
        if any(c.isdigit() for c in word):
            continue

        if len(word) % 2 == 0:
            processed = word[::-1]
        else:
            processed = word.upper()

        if processed not in seen:
            seen.add(processed)
            result.append(processed)

    return result

words = ["hello", "world", "python3", "test", "abc", "hello", "code", "python"]
print("Результат:", analyze_strings_list(words))
```
