```python
#2
filter_set = lambda s: {x for x in s
                       if x > (sum(s) / len(s))
                       and x % 2 == 1
                       and x % 5 != 0}

s = {3, 5, 7, 10, 15, 21, 30, 9}
print("Результат:", filter_set(s))
```

```python
#4
def filter_sets(sets_list):
    result = []

    for s in sets_list:
        if len(s) > 3:
            has_negative = False
            for num in s:
                if num < 0:
                    has_negative = True
                    break

            if not has_negative:
                has_even = False
                for num in s:
                    if num % 2 == 0:
                        has_even = True
                        break

                if has_even:
                    result.append(s)

    return result

sets_list = [{1, 2, 3, 4}, {-1, 2, 3}, {1, 3, 5, 7, 9}, {2, 4, 6, 8}]
print("Результат:", filter_sets(sets_list))
```

```python
#7
symmetric_diff_even = lambda s1, s2: {x for x in (s1 ^ s2) if x % 2 == 0}

s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}
print("Результат:", symmetric_diff_even(s1, s2))
```

```python
#9
def common_elements_all(sets_list):
    if not sets_list:
        return set()

    result = sets_list[0].copy()

    for s in sets_list[1:]:
        new_result = set()
        for elem in result:
            if elem in s:
                new_result.add(elem)
        result = new_result

    return result

sets_list = [{1, 2, 3, 4}, {2, 3, 4, 5}, {2, 3, 4, 6}]
print("Результат:", common_elements_all(sets_list))
```

```python
#14
def top_k_frequent(nums, k):
    freq = {}
    for num in nums:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1

    items = [(count, num) for num, count in freq.items()]

    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            if items[i][0] < items[j][0]:
                items[i], items[j] = items[j], items[i]
            elif items[i][0] == items[j][0] and items[i][1] > items[j][1]:
                items[i], items[j] = items[j], items[i]

    result = set()
    for i in range(min(k, len(items))):
        result.add(items[i][1])

    return result

nums = [1, 2, 2, 3, 3, 3, 4, 4, 5]
k = 2
print("Результат:", top_k_frequent(nums, k))
```

```python
#19
def filter_by_digit_sum(nums):
    result = set()

    for num in nums:
        if num % 2 == 1:
            digit_sum = 0
            temp = abs(num)
            while temp > 0:
                digit_sum += temp % 10
                temp //= 10

            if digit_sum % 2 == 0:
                result.add(num)

    return result

nums = {123, 456, 789, 111, 222, 333}
print("Результат:", filter_by_digit_sum(nums))
```

```python
#24
def union_of_filtered_sets(sets_list):
    result = set()

    for s in sets_list:
        for num in s:
            if num > 10 and num % 2 == 1:
                result.add(num)

    return result

sets_list = [{5, 12, 15, 20}, {7, 13, 17, 22}, {9, 11, 14, 19}]
print("Результат:", union_of_filtered_sets(sets_list))
```

```python
#26
def remove_elements_with_common_digits(s):
    all_digits = {}

    for num in s:
        digits = set(str(abs(num)))
        for digit in digits:
            if digit in all_digits:
                all_digits[digit] += 1
            else:
                all_digits[digit] = 1

    result = set()
    for num in s:
        digits = set(str(abs(num)))
        has_common = False
        for digit in digits:
            if all_digits[digit] > 1:
                has_common = True
                break

        if not has_common:
            result.add(num)

    return result

s = {123, 456, 124, 789, 101}
print("Результат:", remove_elements_with_common_digits(s))
```

```python
#28
def sorted_unique_chars(strings):
    chars = set()

    for s in strings:
        for c in s:
            if c.isalpha():
                chars.add(c.lower())

    result = list(chars)

    for i in range(len(result)):
        for j in range(i + 1, len(result)):
            if result[i] > result[j]:
                result[i], result[j] = result[j], result[i]

    return result

strings = ["Hello", "World123", "Python", "Test!"]
print("Результат:", sorted_unique_chars(strings))
```

```python
#30
def partition_by_sum_parity(s):
    even_sum = set()
    odd_sum = set()

    for num in s:
        digit_sum = 0
        temp = abs(num)
        while temp > 0:
            digit_sum += temp % 10
            temp //= 10

        if digit_sum % 2 == 0:
            even_sum.add(num)
        else:
            odd_sum.add(num)

    return (even_sum, odd_sum)

s = {123, 456, 789, 111, 222, 333}
print("Результат:", partition_by_sum_parity(s))
```

```python
#32
def pairwise_intersections(sets_list):
    if len(sets_list) < 2:
        return []

    result = []

    for i in range(len(sets_list) - 1):
        intersection = set()
        for elem in sets_list[i]:
            if elem in sets_list[i + 1]:
                intersection.add(elem)
        result.append(intersection)

    return result

sets_list = [{1, 2, 3}, {2, 3, 4}, {3, 4, 5}, {4, 5, 6}]
print("Результат:", pairwise_intersections(sets_list))
```

```python
#34
def top_k_smallest_unique(nums, k):
    unique_nums = []
    seen = set()

    for num in nums:
        if num not in seen:
            seen.add(num)
            unique_nums.append(num)

    for i in range(len(unique_nums)):
        for j in range(i + 1, len(unique_nums)):
            if unique_nums[i] > unique_nums[j]:
                unique_nums[i], unique_nums[j] = unique_nums[j], unique_nums[i]

    result = set()
    for i in range(min(k, len(unique_nums))):
        result.add(unique_nums[i])

    return result

nums = [5, 2, 8, 2, 1, 9, 3, 5, 4]
k = 3
print("Результат:", top_k_smallest_unique(nums, k))
```

```python
#36
def all_subsets_of_size_k(s, k):
    elements = list(s)
    result = []

    def generate_subsets(start, current):
        if len(current) == k:
            result.append(set(current))
            return

        for i in range(start, len(elements)):
            current.append(elements[i])
            generate_subsets(i + 1, current)
            current.pop()

    generate_subsets(0, [])
    return result

s = {1, 2, 3, 4}
k = 2
print("Результат:", all_subsets_of_size_k(s, k))
```

```python
#38
def multi_symmetric_difference(sets_list):
    if not sets_list:
        return set()

    result = sets_list[0].copy()

    for s in sets_list[1:]:
        new_result = set()
        for elem in result:
            if elem not in s:
                new_result.add(elem)
        for elem in s:
            if elem not in result:
                new_result.add(elem)
        result = new_result

    return result

sets_list = [{1, 2, 3}, {2, 3, 4}, {3, 4, 5}]
print("Результат:", multi_symmetric_difference(sets_list))
```

```python
#40
def analyze_dict_keys(d):
    chars = set()

    for key in d.keys():
        if isinstance(key, str):
            has_digit = False
            for c in key:
                if c.isdigit():
                    has_digit = True
                    break

            if not has_digit:
                for c in key:
                    if c.isalpha():
                        chars.add(c.lower())

    return chars

d = {'hello': 1, 'world123': 2, 'python': 3, 'test!': 4, 'abc': 5}
print("Результат:", analyze_dict_keys(d))
```
