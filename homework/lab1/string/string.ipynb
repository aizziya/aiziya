```python
#1
def analyze_text(text):
    vowels = set('aeiouy')
    found_vowels = set()

    for char in text.lower():
        if char.isalpha() and char in vowels:
            found_vowels.add(char)

    unique_vowels_count = len(found_vowels)

    words = text.split()
    result_words = []
    seen_words = set()

    for word in words:
        clean_word = ''
        for char in word:
            if char.isalpha():
                clean_word += char

        if len(clean_word) >= 5:
            if clean_word.lower()[0] == clean_word.lower()[-1]:
                if clean_word.lower() not in seen_words:
                    seen_words.add(clean_word.lower())
                    result_words.append(clean_word)

    return (unique_vowels_count, ' '.join(result_words))

text = "Hello Anna level radar World! Anna test"
print("Результат:", analyze_text(text))
```

```python
#2
process_string = lambda s: ' '.join(
    word[::-1] for word in s.split()
    if not any(c.isdigit() for c in word) and len(word) % 2 == 0
)

text = "hello world python3 code test java apple123"
print("Результат:", process_string(text))
```

```python
#3
def top_k_words(text, k):
    text = text.lower()

    clean_text = ''
    for char in text:
        if char.isalpha() or char.isspace():
            clean_text += char

    words = clean_text.split()

    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    items = []
    for word, count in word_count.items():
        items.append((word, count))

    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            if items[i][1] < items[j][1]:
                items[i], items[j] = items[j], items[i]
            elif items[i][1] == items[j][1] and items[i][0] > items[j][0]:
                items[i], items[j] = items[j], items[i]

    result = []
    for i in range(min(k, len(items))):
        result.append(items[i][0])

    return result

text = "apple banana apple orange banana apple grape orange apple"
k = 3
print("Результат:", top_k_words(text, k))
```

```python
#4
filter_words = lambda s: ' '.join(
    word.lower() for word in s.split()
    if sum(1 for c in word if c.isupper()) == 1
    and word[0].isupper() == False
    and word[-1].isupper() == False
)

text = "hello HelLo world Python java JavA test teSt"
print("Результат:", filter_words(text))
```

```python
#5
def compress_text(text):
    if not text:
        return ""

    result = ""
    count = 1
    current = text[0]

    for i in range(1, len(text)):
        if text[i].lower() == text[i-1].lower():
            count += 1
        else:
            if count == 1:
                result += current
            else:
                result += current + str(count)
            current = text[i]
            count = 1

    if count == 1:
        result += current
    else:
        result += current + str(count)

    return result

text = "aaBBcDDD"
print("Результат:", compress_text(text))
```

```python
#6
filter_unique_words = lambda s: [
    word for word in s.split()
    if len(word) >= 4
    and len(set(word)) == len(word)
    and not any(c.isdigit() for c in word)
]

text = "hello world python java tree book cat dog abcde"
print("Результат:", filter_unique_words(text))
```

```python
#7
def palindrome_words(text):
    clean_text = ''
    for char in text:
        if char.isalpha() or char.isspace():
            clean_text += char

    words = clean_text.split()
    palindromes = []
    seen = set()

    for word in words:
        if len(word) >= 3:
            if word.lower() == word.lower()[::-1]:
                if word.lower() not in seen:
                    seen.add(word.lower())
                    palindromes.append(word)

    for i in range(len(palindromes)):
        for j in range(i + 1, len(palindromes)):
            if len(palindromes[i]) < len(palindromes[j]):
                palindromes[i], palindromes[j] = palindromes[j], palindromes[i]
            elif len(palindromes[i]) == len(palindromes[j]) and palindromes[i].lower() > palindromes[j].lower():
                palindromes[i], palindromes[j] = palindromes[j], palindromes[i]

    return palindromes

text = "level radar hello world madam Anna test civic"
print("Результат:", palindrome_words(text))
```

```python
#8
replace_words = lambda s: ' '.join(
    'VOWEL' if word[0].lower() in 'aeiouy' and not any(c.isdigit() for c in word)
    else 'CONSONANT' if not any(c.isdigit() for c in word)
    else word
    for word in s.split()
)

text = "apple banana car123 dog elephant fox"
print("Результат:", replace_words(text))
```

```python
#9
def alternate_case_blocks(text, n):
    result = ""
    block_num = 0

    for i in range(0, len(text), n):
        block = text[i:i+n]
        if block_num % 2 == 0:
            result += block.upper()
        else:
            result += block.lower()
        block_num += 1

    return result

text = "abcdefghijklmnopqrstuvwxyz"
n = 5
print("Результат:", alternate_case_blocks(text, n))
```

```python
#10
count_special_words = lambda s: sum(
    1 for word in s.split()
    if any(c.isdigit() for c in word)
    and not word[0].isdigit()
    and len(word) >= 5
)

text = "hello123 world python3 code test 123abc abc12345 test"
print("Результат:", count_special_words(text))
```

```python
#11
def common_unique_chars(s1, s2):
    result = ""
    seen_in_result = set()

    s2_chars = set()
    for char in s2:
        if char.isalpha():
            s2_chars.add(char)

    for char in s1:
        if char.isalpha() and char in s2_chars and char not in seen_in_result:
            result += char
            seen_in_result.add(char)

    return result

s1 = "hello world 123"
s2 = "world hello 456"
print("Результат:", common_unique_chars(s1, s2))
```

```python
#12
find_special_words = lambda s: [
    word for word in s.split()
    if len(word) > 3
    and word[0].lower() == word[-1].lower()
    and word.lower() != word.lower()[::-1]
]

text = "hello level radar anna test trust start"
print("Результат:", find_special_words(text))
```

```python
#13
def replace_every_nth(text, n, char):
    words = text.split()
    result_words = []

    for word in words:
        if len(word) < 3:
            result_words.append(word)
        else:
            new_word = ""
            for i, c in enumerate(word):
                if (i + 1) % n == 0 and c != ' ' and not c.isdigit():
                    new_word += char
                else:
                    new_word += c
            result_words.append(new_word)

    return ' '.join(result_words)

text = "hello world python code test"
n = 2
char = '*'
print("Результат:", replace_every_nth(text, n, char))
```

```python
#14
filter_complex_words = lambda s: ','.join(
    word for word in s.split()
    if len(set(word.lower())) > 3
    and len([c for c in word.lower() if c in 'aeiouy']) == len(set(c for c in word.lower() if c in 'aeiouy'))
)

text = "hello world python beautiful tree book"
print("Результат:", filter_complex_words(text))
```

```python
#15
def word_pattern_sort(text):
    words = text.split()

    groups = {}
    for word in words:
        length = len(word)
        if length not in groups:
            groups[length] = []
        groups[length].append(word)

    result = []
    sorted_lengths = []
    for length in groups:
        sorted_lengths.append(length)

    for i in range(len(sorted_lengths)):
        for j in range(i + 1, len(sorted_lengths)):
            if sorted_lengths[i] > sorted_lengths[j]:
                sorted_lengths[i], sorted_lengths[j] = sorted_lengths[j], sorted_lengths[i]

    for length in sorted_lengths:
        group = groups[length]

        for i in range(len(group)):
            for j in range(i + 1, len(group)):
                vowels_i = sum(1 for c in group[i].lower() if c in 'aeiouy')
                vowels_j = sum(1 for c in group[j].lower() if c in 'aeiouy')

                if vowels_i < vowels_j:
                    group[i], group[j] = group[j], group[i]
                elif vowels_i == vowels_j and group[i].lower() > group[j].lower():
                    group[i], group[j] = group[j], group[i]

        result.extend(group)

    return result

text = "cat dog elephant bird python apple"
print("Результат:", word_pattern_sort(text))
```
