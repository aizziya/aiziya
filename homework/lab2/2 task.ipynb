```python
import csv
file = open("employees.csv", "r", encoding="utf-8")
reader = csv.DictReader(file)

all_employees = []
department_salaries = {}
department_counts = {}
total_salary = 0
employee_count = 0

for row in reader:
    name = row["name"]
    department = row["department"]
    salary = int(row["salary"])

    all_employees.append({
        "name": name,
        "department": department,
        "salary": salary
    })

    if department in department_salaries:
        department_salaries[department] += salary
        department_counts[department] += 1
    else:
        department_salaries[department] = salary
        department_counts[department] = 1

    total_salary += salary
    employee_count += 1

file.close()

average_salary = total_salary / employee_count

department_averages = {}
best_department = ""
best_avg = 0

for dept in department_salaries:
    avg = department_salaries[dept] / department_counts[dept]
    department_averages[dept] = avg

    if avg > best_avg:
        best_avg = avg
        best_department = dept

top_employee = ""
top_salary = 0

for emp in all_employees:
    if emp["salary"] > top_salary:
        top_salary = emp["salary"]
        top_employee = emp["name"]

high_salary_employees = []
for emp in all_employees:
    if emp["salary"] > average_salary:
        high_salary_employees.append(emp)

print("Средняя зарплата:", average_salary)
print("Средняя по отделам:", department_averages)
print("Лучший отдел:", best_department, "со средней", best_avg)
print("Самый высокооплачиваемый:", top_employee, top_salary)
print("Сотрудников выше средней:", len(high_salary_employees))

output = open("high_salary.csv", "w", encoding="utf-8")
writer = csv.writer(output)
writer.writerow(["name", "department", "salary"])

for emp in high_salary_employees:
    writer.writerow([emp["name"], emp["department"], emp["salary"]])

output.close()
print("Файл high_salary.csv создан!")
```
