```python
import json

file = open("orders.json", "r", encoding="utf-8")
orders = json.load(file)
file.close()

total_revenue = 0
user_orders = {}
item_count = 0
item_frequency = {}
top_user_order = ""
top_order_value = 0

for order in orders:
    order_id = order["order_id"]
    user = order["user"]
    items = order["items"]
    total = order["total"]

    total_revenue += total

    if user in user_orders:
        user_orders[user] += 1
    else:
        user_orders[user] = 1

    item_count += len(items)

    for item in items:
        if item in item_frequency:
            item_frequency[item] += 1
        else:
            item_frequency[item] = 1

    if total > top_order_value:
        top_order_value = total
        top_user_order = user

most_popular_item = ""
max_frequency = 0

for item in item_frequency:
    if item_frequency[item] > max_frequency:
        max_frequency = item_frequency[item]
        most_popular_item = item

summary = {
    "total_revenue": total_revenue,
    "top_user": top_user_order,
    "most_popular_item": most_popular_item,
    "total_orders": len(orders),
    "user_orders": user_orders,
    "total_items_sold": item_count
}

output = open("summary.json", "w", encoding="utf-8")
json.dump(summary, output, indent=2, ensure_ascii=False)
output.close()

print("Общая выручка:", total_revenue)
print("Заказы по пользователям:", user_orders)
print("Всего товаров продано:", item_count)
print("Пользователь с самым дорогим заказом:", top_user_order, "-", top_order_value)
print("Самый популярный товар:", most_popular_item)
print("Файл summary.json создан!")
```
