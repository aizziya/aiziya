```python
import csv
import json

file = open("transactions.csv", "r", encoding="utf-8")
reader = csv.DictReader(file)

fraud_transactions = []
user_transaction_count = {}
fraud_users = set()
suspicious_total = 0

for row in reader:
    user_id = row["user_id"]
    amount = int(row["amount"])

    if user_id in user_transaction_count:
        user_transaction_count[user_id] += 1
    else:
        user_transaction_count[user_id] = 1

    if amount > 500000:
        fraud_transactions.append({
            "user_id": user_id,
            "amount": amount,
            "reason": "сумма > 500000"
        })
        fraud_users.add(user_id)
        suspicious_total += amount

for user in user_transaction_count:
    if user_transaction_count[user] > 3:
        fraud_users.add(user)
report = open("fraud_report.txt", "w", encoding="utf-8")

report.write("Подозрительных транзакций: " + str(len(fraud_transactions)) + "\n")
report.write("Подозрительных пользователей: " + str(len(fraud_users)) + "\n")
report.write("Список пользователей: " + ", ".join(fraud_users) + "\n")
report.write("Общая сумма подозрительных операций: " + str(suspicious_total) + "\n\n")

report.write("Детали подозрительных транзакций:\n")
for trans in fraud_transactions:
    report.write(f"  {trans['user_id']} - {trans['amount']} ({trans['reason']})\n")

report.close()

fraud_users_list = list(fraud_users)

output = open("fraud_users.json", "w", encoding="utf-8")
json.dump(fraud_users_list, output, indent=2, ensure_ascii=False)
output.close()

print("Подозрительных транзакций:", len(fraud_transactions))
print("Подозрительных пользователей:", len(fraud_users))
print("Общая сумма подозрительных операций:", suspicious_total)
print("Файлы fraud_report.txt и fraud_users.json созданы!")
```
