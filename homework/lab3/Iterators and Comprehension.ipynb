```python
#1
result = [x**2 for x in range(1, 21) if x%2 ==0 ]

print(result)
```

```python
#2
matrix = [[1,2,3],[4,5,6],[7,8,9]]
result = [(lambda row: row[0] * row[1] * row[2]) (row) for row in matrix]
print(result)
```

```python
#3
words = ["кот", "машина", "ананас", "дом", 'lekel']
for w in words:
    result = [w for w in words if len(w) > 4 and "а" not in w]

print(result)
```

```python
#4
numbers = [1,2,3,4,5]

result = {x: ('chetnoe' if x%2 ==0 else 'nechetnoe') for x in numbers}

print(result)
```

```python
#5
matrix = [[1,2],[3,4],[5,6]]

result = [num for row in matrix for num in row]

print(result)
```

```python
#6
result = [
"FizzBuzz" if x % 15 == 0 else
"Fizz" if x % 3 == 0 else
"Buzz" if x % 5 == 0 else x
for x in range(1,21)
]

print(result)
```
