```python
#1
def even_numbers(n):
    for i in range(1, n+1):
        if i%2 == 0:
            if i%4==0:
                yield 'kratno 4'
            else:
                yield i
for x in even_numbers(10):
    print(x)
```

```python
#2
def filter_words(words):
    for word in words:
        if len(word) > 4:
            if 'a' in word:
                yield 'c a'
            else:
                yield word

words = ["кот", "мaшина", "aрбуз", "дом"]
for w in filter_words(words):
    print(w)
```

```python
#3
def infinite_numbers():
    n = 1
    while True:
        if n % 3 == 0:
            yield "Fizz"
        if n%5 == 0:
            yield 'Buzz'
        else:
            yield n
        n+=1

```

```python
#4
def squares(n):
    for i in range(1, n+1):
        x = i**2
        if x%2 == 0:
            yield 'even'
        else:
            yield x
for x in squares(5):
    print(x)
```
